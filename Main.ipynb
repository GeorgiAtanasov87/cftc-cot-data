#### Load future

```python
future = cot.get_future('BRENT CRUDE OIL LAST DAY')
df = future.data
df
```

```python
future.load_summary()
```

```python
diff = future.data.diff() * -1
new_names = [(i,'d '+i) for i in diff.iloc[:, :].columns.values]
diff = diff.rename(columns = dict(new_names))
diff = diff.shift(periods=-1)
```

```python
import pandas as pd
import seaborn as sns
cmap=sns.diverging_palette(10, 240, n=9, as_cmap=True)
dff = pd.concat([df, diff], axis=1)

(dff.style.format({'% Long': '{:.0%}',
                       '% Short': '{:.0%}',
                       'Long NC': '{:,}',
                       'Short NC': '{:,}',
                       'Total': '{:,}',
                       'Net Pos': '{:,}',
                       'd % Long': '{:.0%}',
                       'd % Short': '{:.0%}',
                       'd Long NC': '{:,}',
                       'd Short NC': '{:,}',
                       'd Total': '{:,}',
                       'd Net Pos': '{:,}',})
                  .background_gradient(cmap)
        )

```
